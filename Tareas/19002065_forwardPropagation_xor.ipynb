# XOR

Dado el diagrama de red neuronal de una capa oculta en el ejercicio de la diapositiva "clase_MLP" implementar con numpy la etapa de forward propagation para los 4 posibles puntos de la operacion Xor, usar matplotlib para gráficar el resultado de la capa oculta y el Xor original(usando distinto color o marker para diferenciar puntos con y=1 de los puntos con y=0), comparar ambas gráficas y concluir/comentar resultados.
Se debe usar solo código vectorizado.

```python
import numpy as np
import matplotlib.pyplot as plt

print("#########Procesando la función escalón######")

def step(x):
    return (x >= 0).astype(np.float)

x = np.array([1, 0]).reshape(-1, 1)

def ForwardXOR(x): 
  #Primera capa
  P1 = np.array([[20., 20., -10.], [-20., -20., 30.]])
  mul1 = np.matmul(P1, np.vstack((x, np.ones(x.shape[1]))))
  outer1 = step(mul1)

  #Segunda capa
  P2 = np.array([[20, 20, -30]])
  mul2 = np.matmul(P2, np.vstack((outer1, np.ones(outer1.shape[1]))))
  act2 = step(mul2)
   
  return act2, outer1
  
X = (np.array([[0,0], [0,1], [1,0], [1,1]])).T
act2, outer1 = ForwardXOR(X)
```

```python
def predicciones(x1, x2, y):
    abso = np.abs(y-1) < 0.1
    graf = np.logical_not(abso)
    plt.scatter(x1[abso], x2[abso], marker='o', label='1', c='#2ca02c') 
    plt.scatter(x1[graf], x2[graf], marker='X', label='0', c='#d62728')
    plt.legend()
    
x1 = X[0, :]
x2 = X[1, :]

plt.figure(figsize=(17, 6))
plt.subplot(3,1,1)
predicciones(x1, x2, outer1[0, :])
plt.title("capa oculta 1")
plt.subplot(3,1,2)
predicciones(x1, x2, outer1[1, :])
plt.title("capa oculta 2")
plt.subplot(3,1,3)
predicciones(x1, x2, act2.squeeze())
plt.title("Out Layer");
```
