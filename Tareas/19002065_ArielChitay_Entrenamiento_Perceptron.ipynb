```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_name = 'entrenamientoPerceptron.csv'
dataset = pd.read_csv(file_name,encoding='latin-1')
dataset.head()

plt.figure(figsize=(5,5))
plt.scatter(dataset.x1[:], dataset.x2[:], alpha=0.8);
plt.xlabel("Coordenadas en X");
plt.ylabel("Coordenadas en Y");

dataset['bias'] = np.ones(len(dataset))
x = np.array(dataset[['bias', 'x1', 'x2']])
y = dataset['label']
dataset.head()

def train_perceptron(x, y, epochs, lr, m):
    pesos = np.random.standard_normal(size=x.shape[1])
    n = 0        
    while n < epochs:          
        for i in range(len(x)):    
            b = np.dot(x[i], pesos)
            if b >= m:
              yhat = 1 
            else:
              yhat = 0
            for j in range(len(pesos)):             
                pesos[j] = pesos[j] + lr*(y[i]-yhat)*x[i][j]              
        n += 1
    return pesos

perceptron = train_perceptron(x, y, 1000, 0.01, 0)
perceptron

def decision(ejex1):
    return (-(perceptron[0] + perceptron[1]*ejex1)/ perceptron[2])
ejex1 = np.linspace(0,20)
plt.figure(figsize=(5,5))
plt.scatter(dataset.x1[:], dataset.x2[:], alpha=0.8)
plt.plot(ejex1, np.array([decision(x) for x in ejex1]),'r--')
plt.xlabel("Coordenadas en x");
plt.ylabel("Coordenadas en Y");
plt.legend()
plt.show()
```
